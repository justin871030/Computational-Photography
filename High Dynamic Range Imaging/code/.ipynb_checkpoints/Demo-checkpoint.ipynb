```python
import cv2 as cv
import numpy as np
from functools import partial
from HDR_functions import *
```

```python
import matplotlib.pyplot as plt 
from matplotlib.figure import Figure
from matplotlib.ticker import MaxNLocator
```

```python
def show(pic, title=None, subtitle=None):
    plt.figure()
    plt.axis("off")
    if title:
        plt.title(title, fontsize=14)
    if subtitle:
        plt.suptitle(subtitle, fontsize=12)
    plt.imshow(np.moveaxis(np.array([pic[2], pic[1], pic[0]]), 0, -1))
    
def show_images(pic_list):
    plt.figure(dpi=500)
    for i in range(len(pic_list)):
        pic = pic_list[i]
        p = plt.subplot(1, len(pic_list), i+1)
        p.axis("off")
        plt.imshow(np.moveaxis(np.array([pic[2], pic[1], pic[0]]), 0, -1))
    plt.show()
```

## Load Images with different exposure times

```python
TestImage = 'memorial'
src_path = f'../TestImage/{TestImage}'

img_list, exposure_times = LoadExposures(src_path)
radiance = np.zeros_like(img_list[0], dtype=np.float32)
pixel_samples = PixelSample(img_list)
show_images(img_list)
```

## Camera Response Calibration

```python
lambda_ = 50

for ch in range(3):
    response = EstimateResponse(pixel_samples[:,ch,:,:], exposure_times, lambda_)
    radiance[ch,:,:] = ConstructRadiance(img_list[:,ch,:,:], response, exposure_times)
    
plt.imshow(np.log(cv.cvtColor(np.moveaxis(radiance, 0, -1), cv.COLOR_BGR2GRAY)), cmap='jet')
plt.title("radiance", fontsize=16)
plt.colorbar()
plt.show()
```

## White Blance + Global Tone Mapping

```python
gtm_no_wb = GlobalTM(radiance, scale=1)  # without white balance
show(gtm_no_wb, "Without White Blance") 
ktbw = (419, 443), (389, 401)            # range of white color for white balancing
radiance_wb = WhiteBalance(radiance, *ktbw)
gtm_wb = GlobalTM(radiance_wb, scale=1)  # with white balance
show(gtm_wb, "With White Blance")
```

## Pixel Value to Lightness Response

```python
x = np.arange(0,256)
plt.plot(x, np.exp(response[x]))
plt.title("lambda = " + "1e15",fontsize=15)
plt.xlabel("Pixel Value",fontsize=13)
plt.ylabel("Exp g(Pixel Value)",fontsize=13)
plt.show()
```

## Visualization of Camera Response Calibration Results

```python
points = [(97, 203), (463, 399), (77, 76)]
delta_t = np.arange(3,11)
channel = 0
```

```python
points_pix = []
for point in points:
    point_pix = [] 
    for img in img_list[delta_t, channel]:
        point_pix.append(img[point])
    points_pix.append(point_pix)
print(points_pix)
x=np.log(np.array(exposure_times)[delta_t])
for point_pix in points_pix:
    plt.plot(x, point_pix, marker='o', linestyle='dashed')

plt.title("Input pixels",fontsize=15)
plt.xlabel("log Exposure (log t)",fontsize=13)
plt.ylabel("Pixel Value",fontsize=13)
plt.show()
```

```python
x=np.log(np.array(exposure_times)[delta_t])
new_x = []
for point in points:
    new_x.append(x + np.log(radiance[channel][point]))
    
for x, point_pix in zip(new_x, points_pix):
    plt.plot(x, point_pix, marker='o', linestyle='dashed')
plt.title("Estimated response",fontsize=15)
plt.xlabel("log Exposure (log E + log t)",fontsize=13)
plt.ylabel("Pixel Value",fontsize=13)
plt.show()
```

## Local Tone Mapping

```python
TestImage = 'vinesunset'
scale=7
radiance_vinesunset = ReadImg(f'../TestImage/{TestImage}.hdr', -1)
ltm_filter = partial(BilateralFilter, N=15, sigma_s=100, sigma_r=0.8)
ltm_bilateral = LocalTM(radiance_vinesunset, ltm_filter, scale=scale)
show(ltm_bilateral, "scale = "+ str(scale))
```
